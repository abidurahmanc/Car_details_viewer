```python
df.dropna(subset=['Range'],inplace=True)
df['Rating(outof10)'] = df['Rating(outof10)'].replace('na').astype(float)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
categorical_columns = ['Car', 'Style', 'Transmission', 'VehicleType']
for col in categorical_columns:
    df[col] = le.fit_transform(df[col])
```

```python
df
```

```python
import re

def extract_min_price(price_range):
    # Extract the minimum price from the range
    numbers = re.findall(r'\d+\.?\d*', price_range)
    if numbers:
        min_price = float(numbers[0])
        # Convert to lakhs or crores based on the suffix
        if 'Cr' in price_range:
            min_price *= 100  # Convert crores to lakhs
        return min_price
    return None

# Apply the function to the PriceRange column
df['PriceRange'] = df['PriceRange'].apply(lambda x: extract_min_price(x) if pd.notnull(x) else x)

```

```python
def extract_numeric_range(range_value):
    # Extract the numeric values from the range
    numbers = re.findall(r'\d+\.?\d*', range_value)
    if numbers:
        return numbers[0]
    return None

# Apply the function to the Range column
df['Range'] = df['Range'].apply(lambda x: extract_numeric_range(x) if pd.notnull(x) else x)
```
